```python
deaths_cum = pd.read_csv('../../data-truth/RKI/truth_RKI-Cumulative Deaths_Germany.csv', parse_dates=['date'])
```

# By Age

```python
sb.set(font_scale=1.5)
```

# Deaths

### Cumulative

```python
deaths_cum = pd.read_csv('../../data-truth/RKI/by_age/truth_RKI-Cumulative Deaths by Age_Germany.csv', parse_dates=['date'])
```

```python
g = sb.relplot(data=deaths_cum, x="date", y="value", hue="location_name", col="age_group", col_wrap=3, kind='line')
for ax in g.axes.ravel():
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```
